```python
%pip install gmpy2
import gmpy2
from cryptography.hazmat.backends import default_backend
from cryptography.hazmat.primitives.asymmetric import rsa
from cryptography.hazmat.primitives import serialization
```

```python
# Generate a private key.
private_key = rsa.generate_private_key(
    public_exponent=65537,
    key_size=2048,
    backend=default_backend()
)

# Extract the public key from the private key.
public_key = private_key.public_key()

# Convert the private key into bytes (without encryption for simplicity).
private_key_bytes = private_key.private_bytes(
    encoding=serialization.Encoding.PEM,
    format=serialization.PrivateFormat.TraditionalOpenSSL,
    encryption_algorithm=serialization.NoEncryption()
)

# Convert the public key into bytes.
public_key_bytes = public_key.public_bytes(
    encoding=serialization.Encoding.PEM,
    format=serialization.PublicFormat.SubjectPublicKeyInfo
)

# Convert the private key bytes back to a key.
private_key = serialization.load_pem_private_key(
    private_key_bytes,
    backend=default_backend(),
    password=None
)

# Convert the public key bytes back to a key.
public_key = serialization.load_pem_public_key(
    public_key_bytes,
    backend=default_backend()
)


def simple_rsa_encrypt(m, publickey):
    """Encrypt message using RSA public key."""
    numbers = publickey.public_numbers()
    return gmpy2.powmod(m, numbers.e, numbers.n)


def simple_rsa_decrypt(c, privatekey):
    """Decrypt message using RSA private key."""
    numbers = privatekey.private_numbers()
    return gmpy2.powmod(c, numbers.d, numbers.public_numbers.n)

message = 144
ciphertext = simple_rsa_encrypt(message, public_key)
decrypted_message = simple_rsa_decrypt(ciphertext, private_key)

print("Original message:", message)
print("Encrypted message:", ciphertext)
print("Decrypted message:", decrypted_message)
```

```python
from cryptography.hazmat.primitives.asymmetric import rsa, padding
from cryptography.hazmat.primitives import serialization, hashes
import os
```

```python
def generate_keys():
    private_key = rsa.generate_private_key(
        public_exponent=65537,
        key_size=2048
    )
    public_key = private_key.public_key()
    return private_key, public_key

def encrypt_number(public_key, number):
    ciphertext = public_key.encrypt(
        number.to_bytes((number.bit_length() + 7) // 8, byteorder='big'),
        padding.OAEP(
            mgf=padding.MGF1(algorithm=hashes.SHA256()),
            algorithm=hashes.SHA256(),
            label=None
        )
    )
    return ciphertext

def decrypt_number(private_key, ciphertext):
    plaintext = private_key.decrypt(
        ciphertext,
        padding.OAEP(
            mgf=padding.MGF1(algorithm=hashes.SHA256()),
            algorithm=hashes.SHA256(),
            label=None
        )
    )
    return int.from_bytes(plaintext, byteorder='big')

def multiply_encrypted_numbers(private_key, encrypted_num1, encrypted_num2):
    num1 = decrypt_number(private_key, encrypted_num1)
    num2 = decrypt_number(private_key, encrypted_num2)
    return num1 * num2


private_key, public_key = generate_keys()


num1 = 12345
num2 = 6789


encrypted_num1 = encrypt_number(public_key, num1)
encrypted_num2 = encrypt_number(public_key, num2)


multiplied_result = multiply_encrypted_numbers(private_key, encrypted_num1, encrypted_num2)


expected_result = num1 * num2
print(f"Расшифрованный результат умножения: {multiplied_result}")
print(f"Ожидаемый результат: {expected_result}")
```
